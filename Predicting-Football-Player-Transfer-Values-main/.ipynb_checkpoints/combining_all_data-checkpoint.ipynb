```python
import numpy as np
import pandas as pd
```

```python
fbref = pd.read_excel('data/Complete_FBREF.xlsx')
tm = pd.read_excel('data/transfermarkt_scrape_latest.xlsx')
import tools.helpers as th
```

```python
tm
```

```python
fbref
```

```python
df = pd.merge(th.clean_char(tm),fbref,on='Player')
```

```python
list(set([a for a in df[df['League']=='UNKNOWN']['Club']]))
```

```python
df.replace('All information about Leicester (Premier League) ➤ current squad with market values ➤ transfers ➤ rumours ➤ player stats ➤ fixtures ➤ news" name="description"/>',"Leicester City",inplace=True)
df.replace('All information about Newcastle (Premier League) ➤ current squad with market values ➤ transfers ➤ rumours ➤ player stats ➤ fixtures ➤ news" name="description"/>',"Newcastle",inplace=True)
df.replace('All information about Cagliari Calcio (Serie A) ➤ current squad with market values ➤ transfers ➤ rumours ➤ player stats ➤ fixtures ➤ news" name="description"/>','Cagliari',inplace=True)
```

```python
df.loc[df['Club'] == 'Leicester City', ['League']] = 'Premier League'
df.loc[df['Club'] == 'Newcastle', ['League']] = 'Premier League'
df.loc[df['Club'] == 'Cagliari', ['League']] = 'Serie A'
```

```python
df[df['Club']=='Cagliari']
```

```python
df.to_excel('data/final_dataset.xlsx',index=False)
```
